<h1>Power Law Distributions</h1>
We consider here probability distributions of the form 

$p(x) = \frac{\alpha}{x^{\alpha+1}}$

where $x \ge x_{min} \gt 0$ and $\alpha \gt 0$, known as a power law distribution. Such distributions are interesting because they break many significant assumptions used in standard classical statistics. For example, the first moment, or mean, of this distribution is undefined **when $\alpha \le 1$**:

$E(x) = \int_{x_{min}}^{\infty} x \frac{\alpha}{x^{\alpha+1}} = \alpha \frac{x^{1-\alpha}}{1-\alpha}\vert_{x_{min}}^{\infty}$

for $\alpha \lt 1$ and

$E(x) = \alpha \ln x \vert_{x_{min}}^{\infty}$

for $\alpha = 1$, both of which fail to converge in the limit as $x \rightarrow \infty$. Indeed, **all moments, $m \ge \alpha$, of this distribution fail to exist**. In such cases where the mean and standard deviation are not finite, the conditions of the *Central Limit Theorem* are not satisfied.

<h2>Generating Power Law Distributed Data</h2>
In principle, we can generate data following any given probability distribution, $p(x)$ using the following approach. 

1. Formulate the *cummulative distribution function* (CDF), $P(x) = y = \int_{x_{min}}^x p(t)dt$. This is the probability that the random variable with distribution $p(x)$ will be less than or equal to $x$. By definition this function is monotonically increasing and takes values in the interval $[0,1]$.

2. Formulate the *inverse* of the CDF, $x = P^{-1}(y)$. For a given value of $y \in [0,1]$, the inverse CDF gives the value of $x$ where the CDF is equal to $y$. 

3. Generate sample values, $Y = \\{y_1, \ldots, y_N\\}$ from a uniform distribution over $[0,1]$ and compute the inverse CDF for each. This will yield a set of values that are distributed according to the original distribution, $p(x)$

For the power law distribution, the CDF is given by

$P(x) = y = \frac{1}{x_{min}^\alpha} - \frac{1}{x^\alpha}$

Letting $\gamma = \frac{1}{x_{min}^\alpha}$, the inverse CDF is given by

$x = \frac{1}{(\gamma - y)^{1/\alpha}}$

Generating artificial power law data
